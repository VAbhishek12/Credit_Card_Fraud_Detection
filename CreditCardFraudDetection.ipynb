```python

# Standardization...
# converting a data into a standard scale whose mean is 0 and standard deviation is 1.

from sklearn.preprocessing import StandardScaler
SS =StandardScaler()
data[['Amount', 'Time']] = SS.fit_transform(data[['Amount', 'Time']])

data[['Amount', 'Time']].describe()
```

# EDA

```python
# class Distribution..
sns.countplot(x='Class', data=data)
plt.title('Distribution of Classes (0 = Non fraud, 1 = Fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Print actual counts
data['Class'].value_counts()
```

```python
# Amounts for fraud vs non fraud..
plt.figure(figsize=(12, 5))

# Fraudulent Transactions
plt.subplot(1, 2, 1)
sns.histplot(data[data['Class'] == 1]['Amount'], bins=50, color='red')
plt.title('Fraudulent Transactions Amount Distribution')

# Non fraud Transactions
plt.subplot(1, 2, 2)
sns.histplot(data[data['Class'] == 0]['Amount'], bins=50, color='green')
plt.title('Non fraud Transactions Amount Distribution')

plt.tight_layout()
plt.show()
```

```python
# Distribution of transaction time..
plt.figure(figsize=(12, 5))

# Time of Fraud
sns.kdeplot(data[data['Class'] == 1]['Time'], label='Fraud', color='red', fill=True)
sns.kdeplot(data[data['Class'] == 0]['Time'], label='Non Fraud', color='green', fill=True)
plt.title('Distribution of Transaction Time')
plt.xlabel('Time (scaled)')
plt.legend()
plt.show()
```

```python
# Heatmap..
plt.figure(figsize=(15, 10))
# Convert all columns to numeric, coerce errors to NaN
data = data.apply(pd.to_numeric, errors='coerce')
# Drop columns with all NaN values after conversion
data = data.dropna(axis=1, how='all')
corr = data.corr()
sns.heatmap(corr, cmap='coolwarm_r', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Top correlations with the target
top_corr_features = corr['Class'].abs().sort_values(ascending=False)[1:11]
print(top_corr_features)
```

# Apply SMOTE (Synthetic Minority Oversampling Technique)...

```python
# Data Splitting

from sklearn.model_selection import train_test_split

X = data.drop('Class', axis=1)
Y = data['Class']

# split the data into testing and traning..
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
print("Training class distribution:\n", y_train.value_counts())
```

```python
X.head()
```

```python
Y.head()
```

```python
# apply SMOTE..
from imblearn.over_sampling import SMOTE
SM = SMOTE(random_state = 42)
X_train_smote, y_train_smote = SM.fit_resample(X_train, y_train)
print("Training class distribution:\n", y_train_smote.value_counts())
```

```python
sns.countplot(x=y_train_smote)
plt.title("Balanced Class Distribution After SMOTE")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

# Model Building and Evolution...

```python
# importing libraries..

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
# Logistic Regression..

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Train
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_smote, y_train_smote)

# Predict
y_pred_lr = lr.predict(X_test)

# Evaluate
print("🔍 Logistic Regression Evaluation")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr, digits=4))
print("AUC-ROC Score:", roc_auc_score(y_test, y_pred_lr))
```

```python
# Decission Tree..

from sklearn.tree import DecisionTreeClassifier

# Train
dt = DecisionTreeClassifier()
dt.fit(X_train_smote, y_train_smote)

# Predict
y_pred_dt = dt.predict(X_test)

# Evaluate
print("🔍 Decision Tree Evaluation")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_dt))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_dt, digits=4))
print("AUC-ROC Score:", roc_auc_score(y_test, y_pred_dt))
```

```python
# Random Forest..

from sklearn.ensemble import RandomForestClassifier

# Train
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_smote, y_train_smote)

# Predict
y_pred_rf = rf.predict(X_test)

# Evaluate
print("🔍 Random Forest Evaluation")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, digits=4))
print("AUC-ROC Score:", roc_auc_score(y_test, y_pred_rf))
```

# using Neural Networks...

```python
# importing libraries..

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
#Building model..
model = Sequential([
    Dense(32, input_shape=(X_train_smote.shape[1],), activation='relu'),
    Dropout(0.3),
    Dense(16, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])
```

```python
# Compiling the model..
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# model training..
history = model.fit(X_train_smote, y_train_smote,
                    epochs=20,
                    batch_size=32,
                    validation_split=0.2,
                    verbose=1)
```

```python
# model evolution..
y_pred_nn = (model.predict(X_test) > 0.5).astype("int32")

# Metrics
print("🔍 Neural Network Evaluation")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_nn))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_nn, digits=4))
print("AUC-ROC Score:", roc_auc_score(y_test, y_pred_nn))
```

```python
# plotting aaccuracy and loss
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# comparing all models..

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Define a helper to get metrics
def get_metrics(y_true, y_pred, y_prob=None):
    return {
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred, zero_division=0),
        'Recall': recall_score(y_true, y_pred, zero_division=0),
        'F1-score': f1_score(y_true, y_pred, zero_division=0),
        'AUC-ROC': roc_auc_score(y_true, y_prob if y_prob is not None else y_pred)
    }

# Get probabilities for AUC if available
y_prob_lr = lr.predict_proba(X_test)[:,1]
y_prob_dt = dt.predict_proba(X_test)[:,1]
y_prob_rf = rf.predict_proba(X_test)[:,1]
y_prob_nn = model.predict(X_test).ravel()

# Compute metrics
metrics = {
    'Logistic Regression': get_metrics(y_test, y_pred_lr, y_prob_lr),
    'Decision Tree':        get_metrics(y_test, y_pred_dt, y_prob_dt),
    'Random Forest':        get_metrics(y_test, y_pred_rf, y_prob_rf),
    'Neural Network':       get_metrics(y_test, y_pred_nn, y_prob_nn),
}

# Display in a tidy DataFrame
import pandas as pd
df_metrics = pd.DataFrame(metrics).T
print(df_metrics)
```

```python
# roc curve..

import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

plt.figure(figsize=(8,6))

# Compute ROC
for name, y_prob in [('LR', y_prob_lr), ('DT', y_prob_dt),
                     ('RF', y_prob_rf), ('NN', y_prob_nn)]:
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc_score(y_test, y_prob):.3f})')

plt.plot([0,1], [0,1], 'k--', label='Chance')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves Comparison')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve

# Get RF probabilities again
probs = rf.predict_proba(X_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, probs)

# Choose your target recall (e.g., 0.90 for 90%)
target_recall = 0.90
idx = (abs(recall - target_recall)).argmin()

# Define custom threshold
custom_th = thresholds[idx]
print(f"Using threshold = {custom_th:.3f} for ~{target_recall*100:.0f}% recall")
```

```python
y_pred_custom = (probs >= custom_th).astype(int)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

print("🔍 Random Forest with Custom Threshold Evaluation")
print("Threshold:", round(custom_th, 3))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_custom))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_custom, digits=4))
print("AUC-ROC Score (unchanged):", roc_auc_score(y_test, probs))
```
